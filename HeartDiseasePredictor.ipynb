Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection And Processing


```python
# Loading data to the pandas dataframe
heart_data=pd.read_csv('/content/heart.csv')
```

```python
# Printing head of the data
heart_data.head()
```

```python
# Checking numbers of rows and cols
heart_data.shape
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
#Statstical Analysis
heart_data.describe()
```

```python
# checking the distribution of Target 1 fordefected hearts and 0 for Not defected
heart_data['target'].value_counts()
```

```python
# Splitting the Features and Target  axsi 1 for cols and 0 for rows
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting data into Training and Testing data

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
#printing details of train and test data
print(X.shape,  x_train.shape,  x_test.shape,  y_train.shape,  y_test.shape)
```

```python
# Model Training using Logistic Regression
model=LogisticRegression()
```

```python
#Training the LogisticRegression Model with the training data
model.fit(x_train,y_train)
```

Model Evaluation

```python
#Accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on Training data: ',training_data_accuracy)
```

```python
#Accuracy on testing data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on Testing data: ',test_data_accuracy)
```

Building Predictive System

```python
input_data=(62,0,0,138,294,1,1,106,0,1.9,1,3,2)
# change input data to numpy array as originally it is in tuple
input_data_as_numpy_array=np.asarray(input_data)
# reshape the numpy array as we are predicting for only one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction=model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
  print('The person does not have a heart disease')
else:
  print('The person has heart disease')
```
